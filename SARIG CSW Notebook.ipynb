# Accessing SARIG CSW Server with Python

## Description

This Jupyter notebook provides a guide and set of tools for accessing and interacting with the SARIG (South Australian Resources Information Geoserver) CSW (Catalogue Service for the Web) server. The notebook demonstrates how to connect to the SARIG CSW server, query metadata records, and process the retrieved geospatial data. It is intended for users who need to access and manipulate geospatial metadata for applications such as GIS, remote sensing, and spatial data analysis.


## Features

- **CSW Server Connection**: Establishes a connection to the SARIG CSW server.
- **Metadata Querying**: Demonstrates how to query the SARIG CSW server for geospatial metadata.
- **Data Retrieval**: Retrieves and processes geospatial data from the server.
- **Example Queries**: Provides examples of common queries and how to handle the results.

## Example Python


````python
from owslib.csw import CatalogueServiceWeb

# Define the CSW server URL
csw_url = 'https://catalog.sarig.sa.gov.au/csw?service=CSW&version=2.0.2&request=GetCapabilities'

# Create a CSW client
csw = CatalogueServiceWeb(csw_url)

# Perform a query (example: get all records)
csw.getrecords2()
for record in csw.records.values():
    print(record.title)
````
