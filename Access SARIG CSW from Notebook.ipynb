# Access SARIG CSW from Notebook

## Notebook Description

This Jupyter notebook provides a guide and set of tools for accessing and interacting with the SARIG (South Australian Resources Information Geoserver) CSW (Catalogue Service for the Web) server. The notebook demonstrates how to connect to the SARIG CSW server, query metadata records, and process the retrieved geospatial data. It is intended for users who need to access and manipulate geospatial metadata for applications such as GIS, remote sensing, and spatial data analysis.

CSW (Catalog Service for the Web) server is a standardized way to discover and access geospatial metadata. The functions are core to this service:

- GetCapabilities: Provides information about the server's capabilities and supported operations.
- DescribeRecord: Returns metadata describing a specific metadata format.
- GetDomain: Returns domain values for a specified property.
- GetRecords: Returns a list of metadata records matching a specified query.
- GetRecordById: Returns metadata for a specific record identified by its ID.
- GetRepositoryItem: Returns additional information about the repository itself.


## Program Features

- **CSW Server Connection**: Establishes a connection to the SARIG CSW server.
- **Metadata Querying**: Demonstrates how to query the SARIG CSW server for geospatial metadata.
- **Data Retrieval**: Retrieves and processes geospatial data from the server.
- **Example Queries**: Provides examples of common queries and how to handle the results.

## Possible Python Use Cases

### 1. Connect to CSW server


```python
import requests
import json

url = 'https://catalog.uat.sarig.sa.gov.au/csw?Request=GetCapabilities&service=CSW&version=3.0.0&'
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(json.dumps(data, indent=4))
else:
    print(f'Request failed with status code {response.status_code}')
```
