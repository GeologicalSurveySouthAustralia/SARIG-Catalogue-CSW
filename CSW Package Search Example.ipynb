## Example 1 Search Mining And Extraction Of Mineral Resources on CSW

    **Refer to vocabulary page on SARIG catalaugue to understand vocabulary** https://vocabularies.uat.sarig.sa.gov.au/vocab/mer-doc-types

### CSW: This code will connect to the CSW server at https://catalog.uat.sarig.sa.gov.au/csw and search for records related to “Primary Mining And Extraction Of Mineral Resources”.


```python
from owslib.csw import CatalogueServiceWeb

# Create a connection to the CSW server
csw = CatalogueServiceWeb('https://catalog.uat.sarig.sa.gov.au/csw')

# Define the query
query = "Primary Mining And Extraction Of Mineral Resources"

# Use the GetRecords operation to search for records
csw.getrecords2(constraints=[{"cql_text": query}], esn='full')

# Print the results
for rec in csw.records:
    print(csw.records[rec])
```

## Example 2 CKAN API can also be use to find documents
### CKAN : find out Mineral Company report "mesac12583" from SARIG in Python using the requests library.


```python
import requests
import json

# Define the base URL for the CKAN instance
base_url = 'https://catalog.uat.sarig.sa.gov.au/api/action/'

# Define the action and parameters
action = 'package_search'
params = 'fq=name:mesac12583'  # Replace with your specific query

# Construct the full URL
url = base_url + action + '?' + params

# Send a GET request to the CKAN server
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(json.dumps(data, indent=4))
else:
    print(f'Request failed with status code {response.status_code}')
```

### CKAN : Directly download file "2018d036642" to '/download' from SARIG in Python using the requests library.
> **link:**https://catalog.uat.sarig.sa.gov.au/api/action/package_search?fq=name:2018d036642
### The file is public avaliable here:
> **link:**https://demstedpdevaue12.blob.core.windows.net/mesac-public/resources/files/25590/RD3935763_RI%20012%20The%20Geology%20of%20the%20Lyndhurst%20and%20Farina%20Military%20Sheets.PDF

```python
import requests
import os

# Define the base URL for the CKAN instance
base_url = 'https://catalog.uat.sarig.sa.gov.au/api/action/'

# Define the action (might need adjustment based on API)
action = 'package_show'  # Replace if necessary

# Define the package ID
package_id = '2018d036642'  # Replace with your specific package ID

# Construct the full URL
url = base_url + action + '?id=' + package_id

# Send a GET request to the CKAN server
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
  data = response.json()

  # Check if result exists (handle empty results)
  if len(data['result']) > 0:
    # Extract download URL from the first result
    download_url = data['result']['results'][0]['url']

    # Define the download directory
    download_dir = '/download'  # Replace with your specific directory

    # Ensure the download directory exists
    os.makedirs(download_dir, exist_ok=True)

    # Download the file
    filename = os.path.basename(download_url)  # Extract filename from URL
    file_path = os.path.join(download_dir, filename)
    print(f"Downloading {filename} from {download_url}")

    res = requests.get(download_url, stream=True)  # Download in stream mode for larger files
    if res.status_code == 200:
      with open(file_path, 'wb') as f:
        for chunk in res.iter_content(1024):
          f.write(chunk)
      print(f"Download complete: {filename}")
    else:
      print(f"Error downloading file: {res.status_code}")
  else:
    print("No results found for the provided package ID.")
else:
  print(f'Request failed with status code {response.status_code}')

```
